```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')
```

```python
dataset = pd.read_csv(r'E:\Udemy\Machine Learning A-Z\Part 2 - Regression\Section 6 - Polynomial Regression\Position_Salaries.csv')

X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
```

```python
# No need to train and test data as no of observations are very less
```

```python
# Fitting dataset using Linear Regression

from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

```python
# Fitting dataset using Polynomial Regression

from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly,y)
```

```python
# Ploting graph of Linear Regression

plt.scatter(X, y, color='red')
plt.plot(X, lin_reg.predict(X), color='blue', linewidth = 2)
plt.title('Actual vs Bluff (Linear Regression)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

```python
# Ploting graph of Polynomial Regression


plt.scatter(X, y, color='red')
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color='blue', linewidth = 2)
plt.title('Actual vs Bluff (Polynomial Regression)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```
